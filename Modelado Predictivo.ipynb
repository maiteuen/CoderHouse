## Modelado Predictivo del consumo de energía eléctrica, gas y gasóleo en centros de la administración autonómica de la Comunidad de Castilla y León

1. Antecedentes

En la actualidad, la Comunidad de Castilla y León se ha convertido en un referente a nivel nacional en materia energética, siendo la principal comunidad que genera más energía eléctrica renovable que la que consume, pudiendo enviar su excedente a la red eléctrica nacional (Domínguez, 2023). 



En cuanto al gas, Castilla y León, al igual que todo el territorio Español, no es un productor relevante de gas natural, y su consumo depende en gran medida de la red nacional y de importaciones. Sin embargo, la región ha destacado en los últimos años por su producción de biogás y su apuesta por la biomasa, con la esperanza de poder producir el 10% del gas que consume España (Blanco Escalona, 2023).

Si bien este contexto refleja un escenario de cambio, donde la sostenibilidad y la eficiencia energética son prioridades emergentes en la agenda de Castilla y León, es necesario entender lo desde una perspectiva no sólo de la producción, sino también del consumo.

En cuanto al consumo de estas fuentes de energía, asunto totalmente estacional relacionado con la temperatura y el creciente cambio climático, según la Encuesta sobre Hogares y Medio Ambiente del 2018 publicada por el Instituto Nacional de Estadística (INE), los leoneses llegaron a gastar un 20% más en calefacción que el resto de españoles, con una reducción de casi un 20% en el 2022, por el encarecimiento del gas a causa de la invasión en Ucrania y el descenso de temperaturas. 

Por otro lado, el consumo de energía eléctrica en Castilla y León ha experimentado diversos cambios a lo largo de los años: después de un periodo de descensos entre 2011 y 2014, el consumo comenzó a recuperarse en 2015, continuando con un crecimiento constante en los años siguientes. Este crecimiento fue particularmente notable en 2018, cuando Castilla y León, junto con otras regiones como Castilla-La Mancha y Melilla, registró uno de los mayores incrementos en el consumo eléctrico a nivel nacional, sin superar los niveles máximos observados en 2008, antes de la crisis económica​ ​ (Energía de Castilla y León, 2019).

Reducir el consumo de energía eléctrica es importante tanto para disminuir el valor en las facturas de suministro mes a mes, como también para contribuir en la disminución de los gases de efecto invernadero y, con ello, las consecuencias del cambio climático.

2. Objetivo y alcance

El objetivo de esta propuesta de "Productos y Servvicios" para el VIII Concurso de Datos Abiertos de la Junta de Castilla y León es la de presentar un estudio predictivo de consumo energía eléctrica, gas y gasóleo en centros de la administración autonómica de la Comunidad de Castilla y León, accesible a la ciudadanía vía web.

A partir de técnicas estadísticas y programación basadas en ciencia de datos, se predecirá consumo energía eléctrica, gas y gasóleo desde Julio 2024 a Junio 2025, del cual evidentemente todavía no se tienen todos los datos reales.

4. Utilidad

Conocer el consumo de energía eléctrica, gas y gasóleo de centros de la administración autonómica de la Comunidad de Castilla y León puede ser útil tanto para la ciudadanía como tambi´n para la gestión de la misma Junta por varias razones que abarcan aspectos económicos, ambientales, de gestión y de planificación. 

- Este tipo de centros suele manejar presupuestos que son financiados por el dinero de los contribuyentes. Por un lado, al monitorear y gestionar de manera eficiente el consumo energético, se pueden reducir costos operativos, liberando fondos para otras necesidades o proyectos prioritarios. También podrán observarse patrones de consumo que, al ya conocer la potencia contratada en el caso de la energía eléctrica, podría permitir identificar áreas de ineficiencia que podrían estar generando un gasto energético innecesario. 

Con datos precisos sobre el consumo, es posible planificar inversiones en infraestructura energética más eficientes, como la instalación de sistemas de energía renovable, la modernización de equipos, o la implementación de tecnologías de gestión energética.
Desarrollo de Políticas Públicas: Los datos de consumo energético en centros públicos pueden informar la creación de políticas públicas que promuevan la eficiencia energética en todos los sectores, impulsando el uso de tecnologías más limpias y eficientes.


4. Transparencia y Rendición de Cuentas
Transparencia en la Gestión Pública: Monitorear y reportar el consumo energético de los centros públicos contribuye a la transparencia en la gestión de recursos públicos, permitiendo a la ciudadanía evaluar cómo se gestionan los recursos energéticos.
Cumplimiento de Normativas: En muchos lugares, existen normativas que exigen a las instituciones públicas reportar y reducir su consumo energético. Conocer estos datos ayuda a garantizar el cumplimiento de dichas regulaciones.


Por el otro, si bien la ciudadanía ya dispone de estos datos mediante la web de Datos Abiertos de la Junta de Castilla y León, estos datos al no ser gráficos, impide una mejor comprensión. 


5. Educación y Conciencia Pública
Ejemplaridad y Concienciación: Los centros públicos pueden servir de ejemplo en la adopción de prácticas sostenibles y de eficiencia energética. Además, el seguimiento y la divulgación de estos datos pueden concienciar a la población sobre la importancia del ahorro energético.
